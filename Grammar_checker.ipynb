```python
from gramformer import Gramformer
```

```python
gf = Gramformer(models=1)
```

```python
gf.correct("My name are Sawan", max_candidates=1)
```

```python
import gradio as gr
```

```python
def correct_sentence(text):
    result = gf.correct(text, max_candidates=1)
    return list(result)[0]
```

```python
spelling_correction_if = gr.Interface(fn = correct_sentence,
                                      inputs="text",
                                      outputs = "text")
```

```python
spelling_correction_if.launch(debug = True)
```
